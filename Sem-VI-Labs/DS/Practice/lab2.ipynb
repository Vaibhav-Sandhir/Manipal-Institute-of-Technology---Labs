### Q3

```python
data1 = {'Name' : ['Ram', 'Diya', 'Chandan', 'James', 'Alice']}
data2 = {'Maths' : [80.0, 90.0, 77.5, 87.5, 86.5],
         'Physics' : [81.0, 94.0, 74.5, 83.0, 82.5],
         'Chemistry' : [91.5, 86.5, 85.5, 90.0, 91.0],
         'Biology' : [82.5, 83.5, 84.5, 85.0, 93.0]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df_new = pd.concat([df1, df2], axis = 1)
df_new['Total'] = df_new.iloc[:, 1:].sum(axis = 1)
df_new
```

### Q4

```python
name = ['Annie', 'Diya', 'Charles', 'James', 'Emily']
q1 = [8.0, 9.0, 7.5, 8.5, 6.5]
in1 = [11.0, 14.0, 14.5, 13.0, 12.5]
q2 = [9.5, 6.5, 8.5, 9.0, 9.0]
in2 = [12.5, 13.5, 14.5, 15.0, 13.0]

data = {'Name':name, 'Quiz1':q1, 'InSem1':in1, 'Quiz2':q2, 'InSem2':in2}
df = pd.DataFrame(data)
df['Total'] = df.iloc[:, 1:].sum(axis = 1)
means = df.iloc[:, 1:].mean(axis = 0)
row = ["NaN"] + list(means)
df.loc['Mean'] = row
df
```
